### 1.1 擷取數據欄位

```python
# 1.1 只擷取需要的部分(數據)
data = data_raw.iloc[:, 2:] #[取橫排值 , 取縱排值](忽略前兩個[日期]、[測項]，留下純數值)
data
```

### 1.2 轉置成合適架構

```python
# 1.2.1 轉置成合適架構 - 移動
# reset_index https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.reset_index.html
# pd.concat   https://pandas.pydata.org/docs/reference/api/pandas.concat.html
df_raw = pd.DataFrame({})
for day in range( int(len(data)/18) ):
        df_temp=data[18*day:18*(day+1)]         #18個測項為一組移動
        df_temp.reset_index(drop=True, inplace=True)  #重置index
        df_raw = pd.concat([df_raw,df_temp],axis=1)  #axis=1 a每組數據水平向右疊加
df_raw
```

3.1 將資料分為訓練(train)、驗證(validation)

```python
# 3.1 將資料分為訓練資料80%跟測試資料20%
train_dataset = df.sample(frac=0.8, random_state=0)  # frac可設定訓練與測試的比例 (隨機排序)
test_dataset = df.drop(train_dataset.index)
train_dataset
```

## 4.資料標準化


### 4.1 選擇預測的「測項」



```python
# 4.1 把想要預測的「測項」獨立分開出來
train_features = train_dataset.copy()
test_features = test_dataset.copy()
train_labels = train_features.pop('PM2.5') #pop('欄位名稱')取出單一欄位
test_labels = test_features.pop('PM2.5')
train_features
```

### 4.2 建立preprocessing layer 處理標準化

```python
# 4.2 建立preprocessing layer 數據進行訓練前都會先進行標準化
normalizer = tf.keras.layers.Normalization(axis=-1)
normalizer.adapt(np.array(train_features))  #adapt出的平均數和變異數做標準化
```

# 資料視覺化觀察與分析


```python
# # 資料視覺化觀察與分析
# # 各項平均差異過大會導致預測失準(值*權重)
# # 標準化前
train_dataset.describe().transpose()[['mean', 'std']]
```

```python
# # 標準化後
normalized_df=(train_dataset-train_dataset.mean())/train_dataset.std() #(x-平均數)/標準差
normalized_df.describe().transpose()[['mean', 'std']]     #平均數=0 , 標準差=1
# #為什麼平均數沒有等於0? (Floating-point error)
```

# Step 2. 建立模型 線性迴歸 Linear Regression


## 1.建立Model


```python
keras.backend.clear_session() #清除先前訓練紀錄
# 1. 建立Sequential Model
# y= mx + b 產生 1 output

linear_model = tf.keras.Sequential([
    normalizer,        #第一層標準化
    layers.Dense(units=1)   #第二層密集連接層(隱藏層)
])

# Model概況
# normalization 15個輸出
# dense 1輸出(結果)
linear_model.summary()
```

## 2.配置Compile

```python
# 2. 配置Compile
# 配置訓練模型compile,       https://keras.io/zh/models/sequential/#compile
# optimizer    優化器選擇   https://keras.io/zh/optimizers/
# learning_rate  學習速率
# loss       損失函數選擇 https://keras.io/zh/losses/

learning_rate = 0.1   # learning_rate 可稍微調低看看

linear_model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate), #優化器為'Adam'演算法
    loss='mean_absolute_error')      #損失函數為'平均絕對誤差'
```

## 3.訓練Training


```python
%%time
# 3.訓練Training
# https://keras.io/zh/models/model/
# fit訓練，x輸入層，y輸出層
# batch_size預設32，批次輸入的數量
# epochs 完整資料集訓練次數
# verbose日誌顯示模式。0 = 安靜, 1 = 進度, 2 = 每輪一行。
# validation_split將trainig data中的一部分當作驗證資料

batch_size = 32 # batch_size 可大幅度調高看看
epochs = 10   # epochs 可小幅度調高看看

history = linear_model.fit(
    train_features, train_labels,
    batch_size=batch_size,
    epochs=epochs,
    verbose=1,
    validation_split = 0.2)
```

#### 繪圖函式

```python
# 隨epochs增加的loss function變化
def plot_loss(history):
  plt.plot(history.history['loss'], label='loss')
  plt.plot(history.history['val_loss'], label='val_loss')
  #plt.ylim([0, 12])
  plt.xlabel('Epoch')
  plt.ylabel('Error [PM2.5]')
  plt.legend()
  plt.grid(True)
#loss     訓練資料的loss
#val_loss   驗證資料的loss
plot_loss(history)
```

## 4.評估Evaluate

```python
# 收集測試結果
test_results = {}
test_results['linear_model'] = linear_model.evaluate(         # evaluate回傳平均誤差值
    test_features, test_labels, verbose=0)             # 用test資料進行評估
pd.DataFrame(test_results, index=['Mean absolute error [PM2.5]']).T
```

# Step 3. 預測Predict - linear_model


```python
# 圖表呈現預測結果
#test_features測試數據、test_predictions預測結果、linear_model剛剛訓練好的model
test_predictions = linear_model.predict(test_features).flatten()  #flatten 扁平化，依序輸入測試資料
a = plt.axes(aspect='equal')
#繪製散佈圖，test_labels答案，對比test_predictions預測結果
plt.scatter(test_labels, test_predictions)
plt.xlabel('True Values [PM2.5]')
plt.ylabel('Predictions [PM2.5]')
lims = [0, 100]
plt.xlim(lims)
plt.ylim(lims)
_ = plt.plot(lims, lims)

#點越接近對角線，越準確
```

```python
# error distribution 統計相減後的誤差值
error = test_predictions - test_labels
plt.hist(error, bins=25)
plt.xlabel('Prediction Error [PM2.5]')
_ = plt.ylabel('Count')
```

# 深度學習 - Step 2.建立模型 - dnn_model

## 1.建立Model / 2.配置Compile

```python
# 1. 建立Sequential Model
# 迴歸使用深度神經網路(Regression with DNN, Deep Neural Network)
keras.backend.clear_session() #清除先前訓練紀錄
dnn_model = keras.Sequential([
    normalizer,              #標準化
    layers.Dense(100, activation='relu'), #第一層隱藏層(神經元數(輸出數), 激發函數) relu小於0歸零，大於零保留
    layers.Dense(100, activation='relu'), #第二層隱藏層
    layers.Dense(100, activation='relu'), #第三層隱藏層
    layers.Dense(1)            #第四層隱藏層(輸出1結果)
])

# 2. 配置Compile
learning_rate=0.01 # learning_rate 可稍微調低看看

dnn_model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=learning_rate),
    loss='mean_absolute_error')
dnn_model.summary()
```

## 3.訓練Training

```python
%%time
# 3.訓練Training
# fit訓練

batch_size = 1000 # batch_size 可大幅度調高看看
epochs = 10   # epochs 可小幅度調高看看

history = dnn_model.fit(
    train_features, train_labels,
    batch_size=batch_size,
    epochs=epochs,
    verbose=1,
    validation_split=0.2
)
plot_loss(history)
```

## 4.評估Evaluate

```python
# 收集測試結果
# 線性迴歸 vs. 線性迴歸深度學習
test_results['dnn_model'] = dnn_model.evaluate(
    test_features, test_labels,
    verbose=0)
pd.DataFrame(test_results, index=['Mean absolute error [PM2.5]']).T
```
